# Dataset

```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', version=1)
X, y = mnist["data"], mnist["target"]
```

```python
X.shape # X represents 70000 images, each 28x28 pixels
```

```python
y.shape
```

# Show a simple image

```python
import matplotlib.pyplot as plt

some_digit = X.iloc[0] # dataframe
some_digit_image = some_digit.values.reshape(28, 28)  # Reshape into 28x28 image
plt.imshow(some_digit_image, cmap="binary")
plt.axis("off")
plt.show()
```

```python
import numpy as np

# Divide set, mnist is already divided to train and test sets
y = y.astype(np.uint8) # need to convert strings to numpy.uint8
X_train, X_test, y_train, y_test = X[:60000].values, X[60000:].values, y[:60000], y[60000:]
# ??? .values is to get rid of strings????
```

# Classifiers

```python
from sklearn.linear_model import SGDClassifier

y_train_5 = (y_train == 5)
y_test_5 = (y_test == 5)
sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train_5)
```

```python
sgd_clf.predict([some_digit]) #some digit prediction
```

# Cross validation

```python
from sklearn.model_selection import cross_val_score

cross_val_score(sgd_clf, X_train, y_train_5, cv=3, scoring="accuracy") # weak -> not5 returns above 90% because ~10% of images are fives
```

```python
from sklearn.model_selection import cross_val_predict

# let's make prediction and see how many times 5 was confused with something else
# this function returns predictions, not evaluation scores
y_train_pred = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_train_5, y_train_pred)
#[[53892   687]     #negative class (non 5 images). 53892 were correctly classified (true negatives) and 687 were wrongly classified (false positives)
# [ 1891  3530]]    #positive class (5 images). 1891 were wrongly classified (false negatives) and 3530 were correctly classified (true positives)
```

```python
from sklearn.metrics import recall_score, precision_score

precision = precision_score(y_train_5, y_train_pred)
recall = recall_score(y_train_5, y_train_pred)
```

```python
# precision = TP / (TP + FP)
# 3530 / (3530 + 687) = 0.837087977
precision
```

```python
# recall = TP / (TP + FN)
# 3530 / (3530 + 1891) = 0.651171371
recall
```

```python
# harmonic mean f1
my_f1 = 2 / (1/precision + 1/recall)
my_f1
```

```python
from sklearn.metrics import f1_score

# scikit version
scikit_f1 = f1_score(y_train_5, y_train_pred)
scikit_f1
```

```python
from sklearn.metrics import precision_recall_curve

# computing which threshold to use
y_scores = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3, method="decision_function")
precisions, recalls, thresholds = precision_recall_curve(y_train_5, y_scores)
```

```python
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    plt.figure(figsize=(8, 4))  # extra code – it's not needed, just formatting
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision", linewidth=2)
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall", linewidth=2)
    plt.vlines(threshold, 0, 1.0, "k", "dotted", label="threshold")

    # extra code – this section just beautifies and saves Figure 3–5
    idx = (thresholds >= threshold).argmax()  # first index ≥ threshold
    plt.plot(thresholds[idx], precisions[idx], "bo")
    plt.plot(thresholds[idx], recalls[idx], "go")
    plt.ylim(0, 1)
    plt.xlim(-50000, 50000)
    plt.grid()
    plt.xlabel("Threshold")
    plt.legend(loc="center right")
    plt.show()
plot_precision_recall_vs_threshold(precisions, recalls, thresholds)
```

```python
from matplotlib import patches


def plot_precision_vs_recall(precisions, recalls, thresholds):
    plt.figure(figsize=(6, 5))  # extra code – not needed, just formatting

    plt.plot(recalls, precisions, linewidth=2, label="Precision/Recall curve")

    # extra code – just beautifies and saves Figure 3–6
    idx = (thresholds >= threshold).argmax()
    plt.plot([recalls[idx], recalls[idx]], [0., precisions[idx]], "k:")
    plt.plot([0.0, recalls[idx]], [precisions[idx], precisions[idx]], "k:")
    plt.plot([recalls[idx]], [precisions[idx]], "ko",
             label="Point at threshold 3,000")
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.grid()
    plt.legend(loc="lower left")
    plt.show()
plot_precision_vs_recall(precisions, recalls, thresholds)
```

```python
threshold_90_precision = thresholds[np.argmax(precisions >= 0.90)]
threshold_90_precision
```

```python
y_train_pred_90 = (y_scores >= threshold_90_precision)
y_train_pred_90
```

```python
precision_score(y_train_5, y_train_pred_90)
```

```python
recall_score(y_train_5, y_train_pred_90)
```

# ROC CURVE

```python
from sklearn.metrics import roc_curve

#FPR - false positive rate
#TPR - true positive rate (another name for recall)
fpr, tpr, thresholds = roc_curve(y_train_5, y_scores)
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--') # dashed diagonal
plot_roc_curve(fpr, tpr)
```

```python
# one-versus-all strategy (OVA) (also called as one-versus-the-rest) is just simply create a number of classes that
# we want to classify. 0-9 numbers to test means 10 classifiers. We evaluate set with all 10 classifiers and choose
# class with the highest score
```

```python
# one-versus-one (OvO) means creating N*(N-1) / 2 classes and comparing duels. 0 vs 1, 0 vs 2 ...
```

```python
y_train_pred = cross_val_predict(sgd_clf, X_train_scaled, y_train, cv=3)
conf_mx = confusion_matrix(y_train, y_train_pred)
conf_mx
```

```python
plt.matshow(conf_mx, cmap=plt.cm.gray)
plt.show()
# slighty darker color on fives, which means there are few 5s in the dataset or that classifier does not perform well
```

```python
# divide each value in the conf matrix by the number of images of corresponding class
row_sums = conf_mx.sum(axis=1, keepdims=True)
norm_conf_mx = conf_mx / row_sums
```

```python
np.fill_diagonal(norm_conf_mx, 0) # fill diagonal with zeros to keep only the errors
plt.matshow(norm_conf_mx, cmap=plt.cm.gray)
plt.show()
# rows represents each ACTUAL class
# collumns represents each PREDICTED class

# many images gets wrongly classified by 8
# 3 and 5 are confused by both directions
```

# Multilabel classification

```python
from sklearn.neighbors import KNeighborsClassifier

# y_multilabel array has 2 label values ...
y_train_large = (y_train >= 7) # if the digit is (7, 8, 9)
y_train_odd = (y_train % 2 == 1) # if the digit is odd

y_multilabel = np.c_[y_train_large, y_train_odd]
knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train, y_multilabel)

knn_clf.predict([some_digit]) # it outputs 2 labels
# 5 is not larger than 7
# 5 is odd
```

```python
y_train_knn_pred = cross_val_predict(knn_clf, X_train, y_multilabel, cv=3)
f1_score(y_multilabel, y_train_knn_pred, average="macro")
```
